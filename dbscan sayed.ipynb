```python
def calculate_distance(x,y):
    x    = np.array(x)
    y    = np.array(y)
    dist = sqrt(np.power(x-y,2).sum())
    #print dist
    return dist
```

```python
def regionQuery(P, eps, D):	
    neighbourPts = []
    for point in D:
        distance = calculate_distance(P,point)
        if distance <= eps:
            neighbourPts.append(point)
            #print distance
    return neighbourPts
```

```python
def DBSCAN(D, eps, MinPts):
    noise          = []
    visited        = []
    C              = []
    cluster_number = -1
    for point in D:
        #print point
        #visited = np.vstack([visited,point])
        visited.append(point) #marking point as visited
        #print visited
        #print point
        neighbourPts = regionQuery(point, eps, D)
        if len(neighbourPts) < MinPts:
            noise.append(point)
        else:
            C.append([])
            cluster_number+=1
            expandCluster(point, neighbourPts, C, cluster_number,eps, MinPts, D, np.asmatrix(visited))
    #print np.asmatrix(visited)
    print ("no. of clusters: " , len(C))
    print ("length of noise:", len(noise))
    return C
    for cluster in C:
        
        #col =[rand(1),rand(1),rand(1)]
        #print cluster
        plt.scatter([i[0] for i in cluster],[i[1] for i in cluster])
    plt.show()
```

```python
def expandCluster(P, neighbourPts, C, cluster_number,eps, MinPts, D, visited):
    visited = visited.tolist()
    P = list(P)
    C[cluster_number].append(P)
    for point in neighbourPts:
        point = point.tolist()
        if point not in visited:
            visited.append(point) 
            neighbourPts_2 = regionQuery(point, eps, D)
            if len(neighbourPts_2) >= MinPts:
                neighbourPts += neighbourPts_2
        if point not in (i for i in C):
            C[cluster_number].append(point)
```
